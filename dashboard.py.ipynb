```python
import plotly.io as pio
pio.renderers.default = 'notebook'
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
monthly_revenue = (
    df[df['orderstatus'].isin(['Delivered', 'Shipped'])]
    .assign(month=lambda x: x['order_date'].dt.to_period('M').astype(str))
    .groupby('month')['totalprice']
    .sum()
    .reset_index()
)

fig = px.line(monthly_revenue, x='month', y='totalprice', title='Monthly Revenue Trend')
fig.show()
```

```python
fig.update_layout(template='plotly_dark')
fig.update_traces(line=dict(color='white',width=3))
fig.show()
```

```python
top_customers=(
     df[df['orderstatus'].isin(['Delivered', 'Shipped'])]
    .groupby ('customerID')['totalprice']
    .sum()
    .nlargest(5)
    .reset_index()
)         
fig = px.bar(top_customers, x='customerID',y='totalprice', title='top 5 customers by spend')
fig.update_layout(plot_bgcolor='black')
fig.update_traces(marker=dict(color='royalblue'))
fig.update_layout(paper_bgcolor='lightgrey')
fig.show()
```

```python
product_qty = (
    df[df['orderstatus'].isin(['Delivered','Shipped'])]
    .groupby('product')['quantity']
    .sum()
    .reset_index()
)
fig = px.pie(product_qty, names = 'product', values = 'quantity',color_discrete_sequence=px.colors.qualitative.Pastel,
 title = 'product share by quantity sold')
fig.update_layout(plot_bgcolor='grey')
fig.show()
    
```
